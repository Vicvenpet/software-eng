** Data Visualization: Plotting with matplotlib and pylab in python **

```python
import pylab as pl  # alising
#form pylab import *
```

```python
import numpy as np
```

```python
x = np.linspace(-np.pi, np.pi, 1000, endpoint=True)
len(x)
```

```python
sin_x = np.sin(x)
cos_x = np.cos(x)
```

```python
pl.plot(x, sin_x, 'go')
pl.plot(x, cos_x, 'r^')
pl.axis([-4, 4, -1.2, 1.2])
pl.xlabel("x axis ----------->")
pl.ylabel("y axis ----------->")
pl.title("sin_vs_cos graph")
pl.savefig("sin_cos_graph.png")
pl.savefig("sin_cos_graph1.pdf")
pl.savefig("sin_cos_graph2.eps")
pl.show()
```

** Example with matplotlib **

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy
```

```python
x = numpy.arange(-10, 10, 0.01)
len(x)
```

```python
import scipy as sp
```

```python
sinx, cosx = sp.sin(x), sp.cos(x)
```

```python
plt.plot(x, sinx, "go")
plt.plot(x,cosx, "r--")
plt.xlabel("x axis ----->")
plt.ylabel("y axis ----->")
plt.title(" sin(x) and cos(x) function")
plt.savefig("try11.jpeg")
plt.show()
```
