```python
import openpyxl as xl
```

```python
excel_folder = "xl_data"
```

```python
wb = xl.Workbook()
sheet = wb.get_active_sheet()
sheet.title = "experiment_data"
```

```python
names = ["std1", "std2", "std3", "std4"]
roll_no = [1, 2, 3, 4]
eductaion = ["Intermediate", "BSc", "MSc", "BA"]
```

```python
sheet.cell(row=1, column=1).value = "Roll Number"
sheet.cell(row=1, column=2).value = "Name"
sheet.cell(row=1, column=3).value = "Education"
```

```python
i = 2
for r, n, e in zip(roll_no, names, eductaion):
    sheet.cell(row=i, column=1).value = r
    sheet.cell(row=i, column=2).value = n
    sheet.cell(row=i, column=3).value = e
    i = i +1 
```

```python
file_name1 = excel_folder + "/" + "exp_data.xlsx"
print(file_name1)
wb.save(file_name1)
```
